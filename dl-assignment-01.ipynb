```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/lp-5-assignments/Boston.csv')
```

```python
df = df.drop(['Unnamed: 15','Unnamed: 16'],axis=1)
```

```python
df
```

```python
cols = list(df.corr()['MEDV'].sort_values()[-10:].index)
```

```python
import numpy as np
```

```python
df = df[cols]
```

```python
y = np.array(df['MEDV'])
X = np.array(df.drop(['MEDV'],axis=1))
```

```python
from tensorflow.keras import layers
from tensorflow.keras import Sequential
from tensorflow.keras import Model
```

```python
model = Sequential([
    layers.Dense(10, input_shape=(None,9,),activation='relu'),
    layers.Dense(7,activation='relu'),
    layers.Dense(5,activation='relu'),
    layers.Dense(1,activation='linear')
])
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(X_train,y_train,batch_size=8,epochs=200)
```

```python
preds = model.predict(X_test)
```

```python
preds = preds.reshape(-1)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,preds)
```

```python
score
```
